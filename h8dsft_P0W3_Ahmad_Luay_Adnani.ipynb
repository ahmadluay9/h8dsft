### Frequently Purchased Products (Quantity) 

#### Overview

Tabel yang digunakan untuk query ini adalah tabel `order_items` dan tabel `inventory_items`. Primary key dari tabel `order_items` dan Foreign key dari tabel `inventory_items` yaitu pada kolom `product_id`. Di tabel `inventory_items` terdapat kolom `product_category` yang digunakan untuk mendapat jumlah kategori produk yang paling banyak dibeli dengan cara diurutkan berdasarkan jumlah yang paling banyak dibeli. Berikut adalah kategori produk yang paling sering dibeli dengan status `complete` setiap bulannya selama Q1 sampai Q3 tahun 2022 :

````python
# Frequently purchased products based on its quantity with rank #1 each month
rank_quantity = client.query('''
SELECT year, month, category, count_category, rank
FROM (SELECT
        year, month, category, count_category,
        RANK() OVER (PARTITION BY month ORDER BY count_category DESC) AS rank # di rank untuk mengetahui yang paling banyak dibeli
      FROM (SELECT DISTINCT
              EXTRACT(year FROM i.created_at) AS year,
              EXTRACT(month FROM i.created_at) AS month,
              ii.product_category AS category, # jenis kategori
              COUNT(ii.product_category) AS count_category, # jumlah kategori
            FROM `bigquery-public-data.thelook_ecommerce.order_items` AS i
            JOIN `bigquery-public-data.thelook_ecommerce.inventory_items` AS ii
            ON i.product_id = ii.product_id # primary key & foreign key
            WHERE i.status = "Complete" AND
                  i.created_at BETWEEN "2022-01-01" AND "2022-09-30"
            GROUP BY year, month, category) count_table_category
            ) count_table_category_rank
WHERE rank = 1 # rank 1 menunjukkan paling banyak dibeli
ORDER BY month 
''').to_dataframe()

rank_quantity
````

Tabel diatas menunjukkan kategori produk yang paling banyak dibeli (status transaksi: `complete`) setiap bulan selama Q1 sampai Q3 tahun 2022. Kategori **Tops & Tees** merupakan kategori produk yang paling banyak dibeli (status transaksi: `complete`) pada Q1 sampai Q3 tahun 2022.

Primary key dari tabel `order_items` dan Foreign key dari tabel `inventory_items` yaitu pada kolom `product_id`. Di tabel `order_items` terdapat kolom `sale_price` yang dapat menunjukkan total penjualan suatu kategori produk. Di tabel `inventory_items` terdapat kolom `product_category` yang digunakan untuk mengetahui kategori produk yang paling banyak mendapatkan pendapatan. Berikut adalah kategori produk yang paling banyak mendapatkan pendapatan dengan status `complete` setiap bulannya selama Q1 sampai Q3 tahun 2022 :

### Number of Products Sold

Bagian ini memperhitungkan jumlah produk yang terjual (dengan status `complete`) setiap bulannya selama Q1 sampai Q3 tahun 2022. Tabel yang digunakan pada query ini adalah tabel `orders`. Di dalam tabel `orders` terdapat kolom `num_of_items` yang menunjukkan jumlah suatu produk. Berikut adalah tabel jumlah produk yang terjual (dengan status `complete`)setiap bulannya selama Q1 sampai Q3 tahun 2022.

````python
# Jumlah produk yang terjual
item_purchased = client.query('''
SELECT DISTINCT
  extract(year FROM created_at) AS year,
  extract(month FROM created_at) AS month,
  SUM(CASE WHEN  status='Complete' THEN num_of_item ELSE 0 END) AS total_item_complete # kolom `num_of_items` table `orders`
FROM bigquery-public-data.thelook_ecommerce.orders
WHERE created_at BETWEEN "2022-01-01" AND "2022-09-30" 
GROUP BY year,month
ORDER BY year,month ASC;
''').to_dataframe()

item_purchased
````

Tabel diatas memperlihatkan jumlah produk yang terjual (dengan status complete) setiap bulannya selama Q1 sampai Q3 tahun 2022. Perhitungan ini digunakan untuk perhitungan GMV.

## Gross Merchandise Value

Perhitungan gross merchandise value mengacu pada nilai barang yang dijual melalui platform Customer-to-Customer(C2C) atau e-commerce [(Investopedia)](https://www.investopedia.com/terms/g/gross-merchandise-value.asp#:~:text=Gross%20merchandise%20value%20(GMV)%20refers,of%20any%20fees%20or%20expenses..). Rumus untuk perhitungan GMV :



```
GMV = Product Sales Price x Number of Products Sold
```



Berikut adalah perhitungan Gross Merchandise Value.

````python
gmv = pd.DataFrame({'Month': ["January", "February", "March", "April", "May", "June", "July", "August", "September"],
                    'Revenues (Complete)': revenue_all['Total_Sales'],
                    'Item Purchased (Complete)': item_purchased['total_item_complete'],
                    'GMV': revenue_all['Total_Sales'] * item_purchased['total_item_complete']})
gmv.round(2)
````

Tabel diatas menunjukkan nilai GMV cenderung mengalami kenaikkan setiap bulannya yang dapat diartikan :
1. Produk yang dijual banyak (quantity)
2. Banyak produk dengan harga mahal yang terjual (price).
